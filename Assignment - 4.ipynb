#### Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
1. Rank
2. Name
3. Artist
4. Upload date
5. Views

```python
#Connecting to WebDriver
driver = webdriver.Chrome(r'C:\Users\chizz\Downloads\chromedriver_win32.zip\chromedriver.exe')
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
time.sleep(1)
#Maximising the window
driver.maximize_window()

#Scraping the Necessary Info
name = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[2]')
Name = [i.text for i in name]

artist = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[3]')
Artist = [i.text for i in artist]

rank = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[1]')
Rank = [i.text for i in rank]

views = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[4]')
Views = [i.text for i in views]

date = driver.find_elements(By.XPATH,'//*[@id="mw-content-text"]/div[1]/table[2]/tbody/tr/td[5]')
Upload_date = [i.text for i in date]

time.sleep(2)

#Modifying the name column
New_Name = []
for i in Name:
    start_index = i.index('"')
    end_index = i.rindex('"')
    text = i[start_index+1:end_index]
    New_Name.append(text)   

time.sleep(1)
driver.close()
```

```python
#converting Into DataFrame
data = {
    'Rank':Rank,
    'Video Title': New_Name,
    'Uploader/Artist':Artist,
    'Views':Views,
    'Upload Date':Upload_date
}
df = pd.DataFrame(data)
df
```

#### Q5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
1. Repository title
2. Repository description
3. Contributors count
4. Language used

- Note: - From the home page you have to click on the trending option from Explore menu through code.


```python
driver = webdriver.Chrome(r'C:\Users\chizz\Downloads\chromedriver_win32.zip\chromedriver.exe')
driver.get('https://github.com/')
driver.maximize_window()
time.sleep(1)


source = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button').click()
time.sleep(1)
trend = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/ul[3]/li[3]/a').click()
time.sleep(1)

name = driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article/h1/a')
Name = [i.text for i in name]

Desc = driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article/p')
Description =  [i.text for i in Desc]

Url = [i.get_attribute('href') for i in name]
```

```python
Info = []

for url in Url:
    driver.get(url)
    time.sleep(2)
    info = driver.find_element(By.XPATH,'//div[@class="Layout-sidebar"]')
    Info.append(info.text.split('\n'))


Contributors = []
for List in Info:
    for i in List:
        if i.startswith("Contributors"):
            Contributors.append(i.replace('Contributors ',''))
            break
    else:
        Contributors.append('-')
        
        
Languages = []
for List in Info:
    Lang = False
    sublist = []
    for e in List:
        if e == 'Languages':
            Lang = True
        if Lang:
            sublist.append(e)
    Languages.append(sublist)        
    Lang = False
```

```python
#converting Into DataFrame
data = {
    'Name':Name,
    'Description':Description,
    'Contributor Count':Contributors,
    'Languages':Languages
}
df = pd.DataFrame(data)
df
```
