#### Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. 
- You have to scrape the job-title
- job-location
- company_name
- experience_required
You have to scrape first 10 jobs data.

```python
#First Connecting with the Web Driver
driver = webdriver.Chrome(r'C:\Users\chizz\Downloads\chromedriver_win32.zip\chromedriver.exe')
driver.get('https://www.naukri.com/') 
    
# Entering a designation and location as required

designation = driver.find_element(By.CLASS_NAME,'suggestor-input')
designation.send_keys('Data Analyst')
    
location = driver.find_element(By.XPATH,'/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input')
location.send_keys('Bangalore')
    
#Searching for the location
search = driver.find_element(By.CLASS_NAME,'qsbSubmit')
search.click()
```

```python
#Using list comprehension and scraping the required data
 
#For job title 
    
titles = driver.find_elements(By.XPATH,'//a[@class="title fw500 ellipsis"]')
job_title = [i.text for i in titles[:10]]
        
# For Location   
    
location = driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi location"]')
job_loc = [i.text for i in location[:10]]

#For Company name
comp = driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
company_name = [i.text for i in comp[:10]]
    
    
#For Experience req
experience = driver.find_elements(By.XPATH,'//li[@class="fleft grey-text br2 placeHolderLi experience"]//span')
job_exp= [i.text for i in experience[:10]]
```

```python
#Creating a DataFrame 
data = {'Job Title':job_title,'Job Location':job_loc,'Experince required':job_exp,'Company':company_name}
df = pd.DataFrame(data)
df
```

#### Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
1. Brand
2. ProductDescription
3. Price

To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url : https://www.flipkart.com/
2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual.
4. After scraping data from the first page, go to the “Next” Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100 sunglasses.


```python
#First Connecting with the Web Driver
driver = webdriver.Chrome(r'C:\Users\chizz\Downloads\chromedriver_win32.zip\chromedriver.exe')
driver.get('https://www.flipkart.com/') 
```

```python
#Searching for glasses
shades = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
shades.send_keys('sunglasses')    
```

```python
#Clicking the search button
search = driver.find_element(By.CLASS_NAME,'L0Z3Pu')
search.click()
```

```python
Brand_name = []
Product_Description = []
Price = []


for page in range(3):
    brand = driver.find_elements(By.CLASS_NAME,'_2WkVRV')
    Des = driver.find_elements(By.CLASS_NAME,'IRpwTa')
    pricings = driver.find_elements(By.CLASS_NAME,'_3bPFwb')

    for i in brand:
        Brand_name.append(i.text)
    
    for j in Des:
        Product_Description.append(j.text)
    
    for k in pricings:
        Price.append(k.text.split('₹')[1])
    
    next_button = driver.find_element(By.CLASS_NAME,'_1LKTO3')
    next_button.click()
    time.sleep(3)
    
    
```

```python
Brand_name = Brand_name[:100]
Product_Description = Product_Description[:100]
Price = Price[:100]
```

```python
len(Brand_name),len(Price),len(Product_Description)
```

```python
data = {'Brand':Brand_name,'Price':Price,'Description':Product_Description}
```

```python
df = pd.DataFrame(data)
df
```

#### Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpage https://www.azquotes.com/
2. Click on Top Quotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
#First Connecting with the Web Driver
driver = webdriver.Chrome(r'C:\Users\chizz\Downloads\chromedriver_win32.zip\chromedriver.exe')
driver.get('https://www.azquotes.com/') 

#Selecting the Top Quotes option
Top = driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
Top.click()
```

```python
#Scraping top 1000 Quotes
Author = []
Quote = []
Type = []

for page in range(10):
    name = driver.find_elements(By.CLASS_NAME,'author')   
    quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
    kind = driver.find_elements(By.CLASS_NAME,'tags')

    for i in name:
        Author.append(i.text)
    
    for j in quote:
        Quote.append(j.text)
    
    for k in kind:
        Type.append(k.text)
    
    next_button = driver.find_element(By.CLASS_NAME,'next')
    next_button.click()
    time.sleep(3)
    
```

```python
data = {'Author':Author,'Quote':Quote,'Type of Quote':Type}
```

```python
df = pd.DataFrame(data)
```
